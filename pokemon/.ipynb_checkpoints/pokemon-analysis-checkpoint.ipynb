```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas_profiling
```

```python
df = pd.read_csv('datasets/Pokemon.csv')
```

```python
df.head()
```

```python
df.drop(['#'],axis=1,inplace=True)
```

```python
df.Name = df.Name.str.replace(".*(?=Mega)", "")
df.Name = df.Name.str.replace(".*(?=Primal)", "")
```

```python
df = df.set_index('Name')
```

```python
df.head()
```

```python
df_mega = df[df.index.str.contains('Mega ')]
df = df[~df.index.str.contains('Mega ')]
```

```python
print("Max Total: ", df['Total'].idxmax())
print("MAX HP:",df['HP'].idxmax())
print("Max Attack: ", df['Attack'].idxmax())
print("Max Defense: ", df['Defense'].idxmax())
print("Max Sp.Atk: ", df['Sp. Atk'].idxmax())
print("Max Sp.Def: ", df['Sp. Def'].idxmax())
print("Speed: ", df['Speed'].idxmax())
```

```python
sns.countplot(df.Legendary)
```

```python
sns.pairplot(df, vars=['HP','Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed'], hue='Legendary')
plt.show()
```

```python
plt.pie(df['Type 1'].value_counts(), labels=df['Type 1'].value_counts().index, autopct='%1.1f%%')
plt.show()
```

```python
plt.pie(df['Type 2'].value_counts(), labels=df['Type 2'].value_counts().index, autopct='%1.1f%%')
plt.show()
```

```python
df.sort_values(by='Total', ascending=False).drop_duplicates(subset=['Type 1'], keep='first')
```
